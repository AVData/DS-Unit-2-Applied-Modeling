```python
df = pd.read_csv('Traffic_Collision_Data_from_2010_to_Present.csv')
```

```python
import pandas_profiling
```

```python
# Taking a look at all of the columns to see which will be the feature to predict
# from observation, and feature characteristics provided by pandas profiling
# it looks like ther are a few features we can potentially predict
# Features to predict: Victim_Sex

profile_report = df.profile_report(check_correlation_pearson=False,
                                  correlations = {
                                      'pearson':False,
                                      'spearman':False,
                                      'kendall':False,
                                      'phi_k':False,
                                      'cramers':False,
                                      'recoded':False,
                                  },
                                  plot={'histogram': {'bayesian_blocks_bins': False}},
                                  )

profile_report
```
